```python
import pandas as pd
import seaborn as sns
```

# Heart rate

```python
hr = pd.read_csv('../data/fitbit_apr/heartrate_seconds_merged.csv')
hr
```

```python
hr.Time = pd.to_datetime(hr.Time)
```

```python
g = sns.FacetGrid(data=hr, col='Id', col_wrap=3)
g.map_dataframe(sns.scatterplot, x='Time', y='Value', size=10)
```

# Sleep

```python
sleep = pd.read_csv('../data/fitbit_apr/sleepDay_merged.csv')
sleep
```

```python
sleep.SleepDay = pd.to_datetime(sleep.SleepDay)
```

```python
g = sns.FacetGrid(data=sleep, col='Id', col_wrap=3)
g.map_dataframe(sns.scatterplot, x='SleepDay', y='TotalMinutesAsleep')
```
